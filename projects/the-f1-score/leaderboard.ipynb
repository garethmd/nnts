```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = [
    "ETS",
    "TBATS",
    "N-BEATS",
    "WaveNet",
    "Theta",
    "FFNN",
    "DeepAR",
    "PR",
    "(DHR-)ARIMA",
    "CatBoost",
    "Transformer",
    "SES",
    "Prophet",
    "Informer*",
]
points = [439, 423, 385, 372, 351, 338, 327, 321, 304, 273, 236, 178, 177, 18]

# Colors
colors = [
    "#FF5733",
    "#C70039",
    "#900C3F",
    "#581845",
    "#DAF7A6",
    "#FFC300",
    "#FF5733",
    "#C70039",
    "#900C3F",
    "#581845",
    "#DAF7A6",
    "#FFC300",
    "#FF5733",
    "#C70039",
]

# Plotting
fig, ax = plt.subplots(figsize=(10, 8))
bars = ax.barh(models, points, color=colors, edgecolor="black")

# Adding text to bars
for bar in bars:
    width = bar.get_width()
    label_y_pos = bar.get_y() + bar.get_height() / 2
    ax.text(
        width - 25,
        label_y_pos,
        s=f"{width}",
        va="center",
        ha="center",
        color="white",
        fontweight="bold",
    )

# Title and labels
ax.set_title(
    "The F1 Score Forecasting Leaderboard", fontsize=16, fontweight="bold"
)
ax.set_xlabel("Points", fontsize=14)
ax.set_ylabel("Models", fontsize=14)

# Inverting y-axis to have the highest points on top
ax.invert_yaxis()

# Removing top and right spines for better aesthetics
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Adding grid
ax.xaxis.grid(True, linestyle="--", which="major", color="grey", alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```
